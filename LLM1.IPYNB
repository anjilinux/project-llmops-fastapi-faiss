```python
from IPython.display import display
import ipywidgets as widgets

# Input box
text = widgets.Text(
    placeholder='Ask something about LLMOps...',
    description='You:',
    layout=widgets.Layout(width='70%')
)

# Output area
output = widgets.Output()

def handle_query(change):
    query = change['new']
    
    if query.strip() == "":
        return
    
    # Retrieve
    docs = retriever.invoke(query)
    context = "\n".join([d.page_content for d in docs])

    # Prompt
    prompt = f"""
You are a helpful AI assistant.

Answer briefly (2-3 sentences).
Do NOT copy full context.

Context:
{context}

Question: {query}

Answer:
"""

    response = llm(prompt, max_length=120)

    with output:
        print("You:", query)
        print("Bot:", response[0]["generated_text"])
        print("-" * 50)

# Trigger on Enter
text.observe(handle_query, names='value')

display(text, output)
```

```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

# Request format
class Query(BaseModel):
    question: str

@app.post("/chat")
def chat(query: Query):
    user_question = query.question

    # STEP 1: retrieve
    docs = retriever.invoke(user_question)
    context = "\n".join([d.page_content for d in docs])

    # STEP 2: prompt
    prompt = f"""
You are a helpful AI assistant.

Context:
{context}

Question: {user_question}

Answer in one sentence:
"""

    # STEP 3: generate
    response = llm(prompt, max_length=100, do_sample=False)

    return {
        "question": user_question,
        "answer": response[0]["generated_text"]
    }
```
